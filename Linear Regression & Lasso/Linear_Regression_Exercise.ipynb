```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/canada_per_capita_income.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
#data visulaization
plt.xlabel('year')
plt.ylabel('per captia income')
plt.scatter(df.year, df['per capita income (US$)'], color = 'blue')
```

```python
df1 = df.drop('per capita income (US$)', axis = "columns")
df1.head()
```

```python
df2 = df['per capita income (US$)']
df2.head()
```

```python
#model training
model = linear_model.LinearRegression()
model.fit(df1,df2)
```

```python
#predicting the per capita income fo canadian citizens in year 2020
model.predict([[2020]])
```
